#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Work_or_Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
fig, [ax1, ax2] = plt.subplots(1,2, figsize=[12.0,4.0])


# MEDIAN STEPS BY MONTH_ENCODED
#  Dataset and tags:
plt.sca(ax1)
x1 = fitbit.groupby(['Months_encoded'])['Steps'].median()
ax1.set_title('Steps by month', loc='right', fontsize=20, fontweight=300)
bars_1 = ['January','February','March','April','June','July','August','September','October','November','December']
y_pos_1 = [i for i in range(1,13)]

# Create plot:
ax1.plot(x1, color='#DA680F')

# Change the x-axis:
plt.xticks(y_pos_1, bars_1, rotation=90)



# MEDIAN STEPS BY WORK_OR_WEEKEND
#  Dataset and tags:
plt.sca(ax2)
x2 = fitbit.groupby(['Work_or_Weekend'])['Steps'].median()
ax2.set_title('Steps by Weekday', loc='right', fontsize=20, fontweight=300)
bars_2 = ('Weekdays', 'Weekends')
y_pos_2 = np.arange(len(bars_2))

# Create plot:
plt.bar(y_pos_2, x2, color=['#ff9478','#d91e18'])

# Create names on the x-axis:
plt.xticks(y_pos_2, bars_2)

plt.show()

# Save your figures in a folder called figures in your repo. 
fig.savefig('figures/stepvstime.png')
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# All in one:
count = 0
for col in fitbit[['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']]:
    sns.scatterplot(x=col, y='Steps', data=fitbit, color= colors[count], alpha=0.7)
    count +=1
```

```python
# Three different subplots:

# Set environment:
f = plt.figure()
colors = ['#c7ede4', '#a35b5b', '#af9ab2', '#820b8a', '#672a4e']
count = 1

# Looping over columns and colors. Setting subplot size:
for col in fitbit[['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']]:
    f.add_subplot(1, 3, count)
    f.set_size_inches(11.7, 8.27)
    sns.scatterplot(x=col, y='Steps', data=fitbit, color= colors[count], alpha=0.9)
    count +=1

# Adjusting separation:
plt.subplots_adjust(left=0, right=0.9, wspace=0.7)
plt.xticks(rotation=90)
```

#### Plot the correlation matrix using seaborn.

```python
# Computing the correlation matrix and assigning it to the variable c:
c = titanic.corr()

# Choosing a color palette:
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Making the heatmap:
sns.heatmap(c, cmap=cmap, annot=True,
           vmin=-1, vmax=1, center=0)
```
